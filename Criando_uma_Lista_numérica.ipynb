```python
#neste caso sera criada uma lista e o programa percorrera a todos os numero de 1 a 10 elevando os mesmo ao quadrado 
quadrado = []
for value in range (1,11):
  quadrad = value**2
  quadrado.append(quadrad)
print(quadrado)  
```
